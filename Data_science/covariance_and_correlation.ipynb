# What is Correlation ?
- Correlation analysis is a method of statistical evalution used tostudy the strength of a relationship between two, numerically measured, continuous variables.

- Range of correlation is between -1 and 1.

   - Correlation = Cov(x,y) / σxσy
   - Where σx and σy are the standerd deviattion of x and y respectively.
