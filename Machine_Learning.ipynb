# What is Machine Learning.
- Machine learning is a subset of artificial intelligence that involves the use of algorithum and statistics to enable machines to learn from data, without using explicit programming.

# How does Machine learning work?
- A machine learning system learns from historical data, builds the prediction models, and whenever it receives new data, oredicts the output for it.

# Classification Of Machine Learning.

- Supervised Learning: In this type of learning, the algorithum is trained on label data. The algorithm learns the relationship between the input data and the output data. The goal of supervised learning is to predict.

- Unsupervised Learning: In this type of learning, the algorithm is trained on unlabeled data. The algorithm learns the relationship between the input data. The goal of unsupervised learning is to discover . 

- Reinforcement Learning: In this type of learning, the algorithm is trained on a reward signal. The algorithm learns the relationship between the input data and the output data. The goal of reinforcement learning is to maximize the reward.



# Advantages of machine Learning.
- **Scalability**: Machine learning algorithms can handle large amounts of data and scale up to meet the needs of complex systems.

- No human intervation neeeded(automation).

- Handling multi-dimensional and multi-variety data.

- Wide applications.


# Disantages of machine learning.
- **Interpretability**: Machine learning models can be difficult to interpret, making it challenging to understand why a particular decision was made.

- Time and resources: Training a machine learning model can be time-consuming and require significant computational resources.

- High error-susceptibility: Machine learning models can be prone to errors, especially if the data used to train them is biased or incomplete.
